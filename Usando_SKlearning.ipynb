```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np
import matplotlib.pyplot as plt

# Dados (Peso, Pressão Arterial)
data = {
    'Peso': [78.5, 83.5, 88.0, 95.7, 88.9, 99.8, 85.3, 85.3, 93.9, 75.7, 98.4],
    'Pressão Arterial': [132, 143, 153, 162, 154, 168, 137, 149, 159, 128, 166]
}

# Convertendo os dados para um DataFrame
df = pd.DataFrame(data)

# Redimensionando X para ter uma dimensão extra (necessário para regressão linear com uma única variável)
X = df[['Peso']]
y = df['Pressão Arterial']

# Dividindo os dados em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# Criando o modelo de regressão linear
model = LinearRegression()

# Treinando o modelo com os dados de treino
model.fit(X_train, y_train)

# Fazendo previsões com os dados de teste
y_pred = model.predict(X_test)

# Avaliando o modelo
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Exibindo os coeficientes da regressão e o erro com 2 casas decimais
print(f'Coeficiente da Regressão: {model.coef_[0]:.2f}')
print(f'Intercepto: {model.intercept_:.2f}')
print(f'Erro Quadrático Médio (MSE): {mse:.2f}')
print(f'Raiz do Erro Quadrático Médio (RMSE): {rmse:.2f}')

# Fazendo uma previsão para um novo valor de peso
new_data = [[75.0]]  # Exemplo de novo peso
predicted_pressure = model.predict(new_data)
print(f'Previsão de Pressão Arterial com base no Peso 75.0: {predicted_pressure[0]:.2f}')

# Plotando o gráfico
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', label='Dados Reais')
plt.plot(X, model.predict(X), color='red', label='Linha de Regressão')
plt.title('Regressão Linear: Pressão Arterial em Função do Peso')
plt.xlabel('Peso (kg)')
plt.ylabel('Pressão Arterial (mmHg)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Peso para o qual queremos prever a pressão arterial
peso_110kg = [[110.0]]

# Fazendo a previsão usando o modelo treinado
pressao_predita_110kg = model.predict(peso_110kg)
print(f'Previsão de Pressão Arterial para Peso 110 kg: {pressao_predita_110kg[0]:.2f} mmHg')
```
